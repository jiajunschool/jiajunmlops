# Task 1

```python
import pandas as pd
import matplotlib as plt

# Load the dataset
file_path = '02_mushroom_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
data
```

```python
# Remove columns with all missing values if any
cleaned_data = data.dropna(axis=1, how='all')

# Display the cleaned dataset's shape and first few rows to confirm cleaning
cleaned_shape = cleaned_data.shape
cleaned_data.head(), cleaned_shape
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

## Visualization 1: Histogram of mushroom class distribution


```python
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data=cleaned_data)
plt.title('Distribution of Mushroom Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

## Visualization 2: Bar plot of cap color distribution


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cap-color', data=cleaned_data)
plt.title('Distribution of Mushroom Cap Colors')
plt.xlabel('Cap Color')
plt.ylabel('Count')
plt.show()
```

## Visualization 3: Distribution of odor across mushroom classes

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='odor', hue='class', data=cleaned_data, palette='coolwarm')
plt.title('Distribution of Odor Across Mushroom Classes')
plt.xlabel('Odor')
plt.ylabel('Count')
plt.legend(title='Class')
plt.show()
```

# Task 2 (Data Transformation, Pycaret Initilization, MLFLOW)

```python
from pycaret.classification import *

# Prepare the dataframe with top 5 features + target
selected_features = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor']
target = 'class'
df_selected = cleaned_data[selected_features + [target]]

# Initialize the PyCaret setup
clf_setup = setup(data=df_selected, target=target, session_id=123, verbose=False, preprocess=True)
```

```python
df_selected
```

```python
best_model = compare_models()
```

```python
# If you want to create a specific model, e.g., Random Forest
from pycaret.classification import create_model
model = create_model('rf')  # 'rf' is the ID for RandomForest
```

```python
from pycaret.classification import tune_model
tuned_model = tune_model(model)
```

```python
from pycaret.classification import evaluate_model
evaluate_model(tuned_model)
```

```python
import pandas as pd

# Example user input for the features
user_input = {}
user_input['cap-shape'] = input('Enter cap-shape: ')
user_input['cap-surface'] = input('Enter cap-surface: ')
user_input['cap-color'] = input('Enter cap-color: ')
user_input['bruises'] = input('Enter bruises: ')
user_input['odor'] = input('Enter odor: ')

# Convert user input into a DataFrame
user_input_df = pd.DataFrame([user_input])

# Assuming `tuned_model` is your trained and tuned model from PyCaret
predictions = predict_model(tuned_model, data=user_input_df)

# Assuming 'Label' is the prediction and 'Score' is the confidence score
# Adjust these column names based on the actual output
if 'prediction_label' in predictions and 'prediction_score' in predictions:
    print("Prediction:", predictions['prediction_label'].iloc[0])
    print("Prediction Score:", predictions['prediction_score'].iloc[0])
else:
    # If the expected columns are not present, print the available columns to help diagnose
    print("The expected columns ('Label', 'Score') are not in the prediction output. Available columns:", predictions.columns)
```
